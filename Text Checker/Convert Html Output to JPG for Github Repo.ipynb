```python
import os
import imgkit

# Path to wkhtmltoimage executable
config = imgkit.config(wkhtmltoimage='C:/Program Files/wkhtmltopdf/bin/wkhtmltoimage.exe')

def convert_html_to_jpg(directory):
    if not os.path.exists(directory):
        print(f"Directory {directory} does not exist.")
        return

    files = [f for f in os.listdir(directory) if f.endswith('.html')]

    if not files:
        print(f"No HTML files found in directory {directory}.")
        return

    for file in files:
        html_path = os.path.join(directory, file)
        jpg_path = os.path.join(directory, f"{os.path.splitext(file)[0]}.jpg")
        
        try:
            imgkit.from_file(html_path, jpg_path, config=config)
            print(f"Converted {file} to {jpg_path}")
        except Exception as e:
            print(f"Failed to convert {file}: {e}")

# Example usage
directory_path = './Outputs/zz472cp8582_jpg'
convert_html_to_jpg(directory_path)
```
